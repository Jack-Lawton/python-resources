# Matplotlib

Matplotlib is the visualisation library of choice for most python developers. It is highly flexible and able to plot just about anything required within the world of data science.

## Basics

```python

# Import matplotlib.pyplot (convention is to import as plt)
import matplotlib.pyplot as plt

# Define x and y data sets
x = range(5)
y = range(5)

# Plot data
plt.plot(x, y)

# Display complete visual
plt.show()
```

## Chart Labels

```python

# Plot data
plt.plot(x, y)

# Set labels for x-axis, y-axis and title
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Chart Title")

# Display visual
plt.show()
```

## Multiple Series

The `plot` function may be called multiple times before the `show` function, allowing mutliple lines to be added to a single chart.

```python

# Additional data
x2 = range(5)
y2 = range(5)[::-1]

# Plot data
plt.plot(x, y)
plt.plot(x2, y2)

# Display visual
plt.show()
```

It's also useful to utilise the legend and label functionality to differentiate series.

```python

# Plot data with labels
plt.plot(x, y, label="Origional Data")
plt.plot(x2, y2, label="New Data")

# Enable the legend plot element
plt.legend()

# Display visual
plt.show()
```

Remember, you can call `plt` as many times as required, even in a loop!

```python

# Loop to generate and plot some new data
for i in range(50):
    tmp_y = [j * i / 50 for j in y]
    plt.plot(x, tmp_y)

# Display visual
plt.show()
```

Note how colours are automatically chosen in a loop.

## Colours

You can specify line colours in matplotlib very easily using a named parameter (be careful, matplotlib always uses the american spelling of colour - "color").


```python

# Plot red line
plt.plot(x, y, color="red")

# Display visual
plt.show()
```


There are a wide range of colours available, as [documented here](https://matplotlib.org/stable/gallery/color/named_colors.html).

![Matplotlib Colours](https://matplotlib.org/stable/_images/sphx_glr_named_colors_003.png "Matplotlib Colours")


For complete flexibility, you can also provide colours as hex codes or RGB values

```python

# Plot data
plt.plot(x, y, color="#2abd84")
plt.plot(x2, y2, color=(1, 0, 0.5))

# Display visual
plt.show()
```


## Line Styles

A line's style and width can be configured using the `linestyle` and `linewidth` (defaults to 1) parameters.


```python

# Plot a dotted line
plt.plot(x, y, linestyle=":", linewidth=3)

# Display visual
plt.show()
```


Several linestyles are available as [documented here](https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html).

![Line Styles](https://matplotlib.org/3.0.3/_images/sphx_glr_line_styles_reference_001.png "Line Styles")

Additionally, there is the `" "` linestyle for an invisible line. This is useful (in combination with markers) for creating scatterplots.


## Markers

Another way to help differentiate lines is by using different markers. Markers are symbols which sit on each point of a line.


```python

# Plot a line with markers
plt.plot(x, y, marker="o")

# Display visual
plt.show()
```


Many markers are available and [defined here](https://matplotlib.org/stable/api/markers_api.html). However, some of the more useful are as follows:

* `o` - Circle
* `x` - Cross
* `^` - Triangle (up)
* `v` - Triangle (down)
* `s` - Square
* `,` - Single Pixel


# Pandas

Matplotlib supports x and y values as any iterable data type, however, often when working with data you will find yourself using [pandas](https://pandas.pydata.org/).

Pandas is a common python library for data analysis and manipulation.



To demonstrate using pandas with matplotlib, let's use a real-world example. We can take the excel file available from [this freedom of information request](https://www.whatdotheyknow.com/request/distance_between_adjacent_underg) and read it in using pandas.

```python
# Import pandas (convention to import as pd)
import pandas as pd
# Display function to view data (jupyter notebooks only)
from IPython.display import display

# Read excel with pandas
df = pd.read_excel("../data/Inter station database.xls", sheet_name="Runtime", skiprows=1)

# Display
display(df)
```

This dataset contains running times and distances between stations for all London Underground lines, a rich dataset we can use to test different types of visualisation.

Notes:
* Although I will be using a few different pandas techniques in this lesson, I will not be going into detail about pandas.
* If you are using Databricks, you may have [PySpark dataframes](https://spark.apache.org/docs/latest/sql-programming-guide.html), whilst it is possible to visualise these directly with matplotlib, I recommend converting them to pandas dataframes using `df = df.toPandas()`. For larger datasets, Databricks supports a [more scaleable inline visualisation](https://docs.databricks.com/notebooks/visualizations/index.html).

For example, lets use it to plot station distance against running time.

```python

# Plot (it can be useful to spread this over multiple lines for code readability)
plt.plot(df["Distance (Kms)"], 
         df["Un-impeded Running Time (Mins)"])

# Axis labels
plt.xlabel("Distance")
plt.ylabel("Running Time")

# Display visual
plt.show()
```

## Extension

Break up the data by line, giving each one its own colour.


## Solution

```python

# Create a lookup for lines and colours
lines_and_colour = [
    ("Bakerloo ", "brown"),
    ("Central ", "red"),
    ("Circle ", "yellow"),
    ("District", "green"),
    ("East London ", "orange")
]

# Loop all lines
for line, colour in lines_and_colour:
    # Filter for just this line
    temp_df = df[df["Line"] == line]
    # Plot
    plt.plot(temp_df["Distance (Kms)"], 
             temp_df["Un-impeded Running Time (Mins)"],
             linestyle=" ",  # Invisible line style
             marker="x",     # Cross markers
             color=colour)   # Colour

# Axis labels
plt.xlabel("Distance")
plt.ylabel("Running Time")

# Display visual
plt.show()

```

# Bar Charts

While line and scatter plots are fairly straight forward, bar charts are a little more complicated. For an example, let's plot number of stations per line.


```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Calculate stations per line

stations_per_line = {
    line: len(df[df["Line"] == line])
    for line in set(df["Line"])
}

# Get x locactions for each bar
x_locs = range(len(stations_per_line))

# First plot the bars using values
plt.bar(x_locs, stations_per_line.values())

# Add bar labels using keys
plt.xticks(x_locs, stations_per_line.keys())

# Label axes
plt.xlabel("Line")
plt.ylabel("N. Stations")

plt.show()
```

To make each line a different colour, we must call bar separately, a pain, but the principle is similar to stacked and grouped bar charts.

```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Loop lines
for i, stations in enumerate(stations_per_line.values()):
    # Plot (single bar)
    plt.bar(i, stations)
    
# Add bar labels using keys
plt.xticks(x_locs, stations_per_line.keys())

# Label axes
plt.xlabel("Line")
plt.ylabel("N. Stations")

plt.show()
    
```

## Stacked Bar Charts

To add a stacked bar chart, we do this the same way but must specify the 'bottom' of the second bar


```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Calculate average travel times

average_times = {
    line: df[df["Line"] == line]["Un-impeded Running Time (Mins)"].mean()
    for line in set(df["Line"])
}

# Get x locactions for each bar
x_locs = range(len(stations_per_line))

# First plot the bars using values
plt.bar(x_locs, stations_per_line.values(), label="Stations")
plt.bar(x_locs, average_times.values(), bottom=list(stations_per_line.values()), label="Times")

# Add bar labels using keys
plt.xticks(x_locs, stations_per_line.keys())

# Label axes
plt.xlabel("Line")
plt.ylabel("N. Stations")

# Enable legend
plt.legend()

plt.show()

```

## Grouped Bar Charts

The stacked bar chart isn't always the most useful, as above the scales are too different - we can use a grouped bar chart and second axis to get around this. 

```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Plot first bars as normal, but with a reduced width
plt.bar(x_locs, stations_per_line.values(), width=0.35, label="Stations")
# Shift the second bars along by a set amount (in this case the width)
bar2_locs = [loc + 0.35 for loc in x_locs]
plt.bar(bar2_locs, average_times.values(), width=0.35, label="Times")

# Add bar labels using keys
plt.xticks(x_locs, stations_per_line.keys())

# Label axes
plt.xlabel("Line")
plt.ylabel("N. Stations")

# Enable legend
plt.legend()

plt.show()
```

## Second Axis

To create a second axis, we need to use the `ax` (axis) object created by the `subplots` method to call all the same functions which we typically use with `plt`, for example:

```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Plot first bars as normal, but with a reduced width
ax.bar(x_locs, stations_per_line.values(), width=0.35, label="Stations")
# Shift the second bars along by a set amount (in this case the width)
bar2_locs = [loc + 0.35 for loc in x_locs]
ax.bar(bar2_locs, average_times.values(), width=0.35, label="Trains")

# Add bar labels using keys
ax.set_xticks(x_locs)
ax.set_xticklabels(stations_per_line.keys())

# Label axes
ax.set_xlabel("Line")
ax.set_ylabel("N. Stations")

# Enable legend
plt.legend()

plt.show()
```

Some of the functions are now prefixed `set_`, like `set_xlabel`, but note that the functionality is typically similar.

Now we can create a second axis and use that to plot our second bars.

```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Plot first bars as normal, but with a reduced width
ax.bar(x_locs, stations_per_line.values(), width=0.35, label="Stations")

# Create a second axis
ax2 = ax.twinx()

# Plot average times against ax2
bar2_locs = [loc + 0.35 for loc in x_locs]
ax2.bar(bar2_locs, average_times.values(), width=0.35, color="orange", label="Time")
# Note we need to speficy the colour now, or both bars will be blue

# Add bar labels using keys
ax.set_xticks(x_locs)
ax.set_xticklabels(stations_per_line.keys())

# Label axes
ax.set_xlabel("Line")
ax.set_ylabel("N. Stations")
ax2.set_ylabel("Average Time")

# Legend
ax.legend()

plt.show()
```

Note that the legend only works for one axis, by default. To correct this, we need to use the reference objects returned from `ax.bar`

```python

# Set plot size first (you may need to play with the numbers to suit your system)
fig, ax = plt.subplots(figsize=(18, 6))

# Plot first bars as normal, but with a reduced width
p1 = ax.bar(x_locs, stations_per_line.values(), width=0.35)

# Create a second axis
ax2 = ax.twinx()

# Plot average times against ax2
bar2_locs = [loc + 0.35 for loc in x_locs]
p2 = ax2.bar(bar2_locs, average_times.values(), width=0.35, color="orange")
# Note we need to speficy the colour now, or both bars will be blue

# Add bar labels using keys
ax.set_xticks(x_locs)
ax.set_xticklabels(stations_per_line.keys())

# Label axes
ax.set_xlabel("Line")
ax.set_ylabel("N. Stations")
ax2.set_ylabel("Average Time")

# Legend
ax.legend([p1, p2], ["Stations", "Time"])

plt.show()
```

# Multiple plots

# Polygons

Sometimes it may be useful to plot shapes. I recomend using polygon objects from [Shapely](https://pypi.org/project/Shapely/). In this example I also use the [Descartes](https://pypi.org/project/descartes/) to plot them.

N.B. For installing shapely on windows, I highly recomend [this source](https://www.lfd.uci.edu/~gohlke/pythonlibs/#shapely) for built wheels.


```python

# Import the polygon object from Shapely
from shapely.geometry.polygon import Polygon

# Import descartes to plot the polygon
from descartes import PolygonPatch

# Setup plot
fig, ax = plt.subplots(figsize=(18, 6))

# Note that polygons will not automatically set your limits
# Set limits manually
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 2.5)

# Create a polygon
my_poly = Polygon([(0, 0), (0, 2), (1, 1),
                   (2, 2), (2, 0), (1, 0.8), (0, 0)])

# Create a polygon patch (using descartes)
my_patch = PolygonPatch(my_poly)

# Plot polygon
ax.add_patch(my_patch)

# Show
plt.show()
```

Note that polygon visual properties are set at the patch level, as [documented here](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Polygon.html).

```python
# Setup plot
fig, ax = plt.subplots(figsize=(18, 6))

# Set limits manually
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 2.5)

# Create a polygon patch (using descartes)
my_patch = PolygonPatch(my_poly, color="red")

# Plot polygon
ax.add_patch(my_patch)

# Show
plt.show()
```
